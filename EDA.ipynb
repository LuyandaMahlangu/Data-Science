```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
data = pd.read_csv('iris.csv')
data
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Checking for duplicates 
data.duplicated().sum()
```

```python
#Histograms for numerical columns
plt.figure(figsize=(10, 8))
data.drop('species', axis = 1).hist(bins=20)
plt.suptitle('Distribution of Sepal and Petal Measurements')
plt.tight_layout()
plt.show()
```

```python
#Checking  for data types 
data.info
```

```python
# KDE Plots for better visualization 
plt.figure() 
sns.kdeplot(data['sepal_length'], shade=True, label='Sepal Length') 
sns.kdeplot(data['sepal_width'], shade=True, label='Sepal Width') 
sns.kdeplot(data['petal_length'], shade=True, label='Petal Length') 
sns.kdeplot(data['petal_width'], shade=True, label='Petal Width') 
plt.title('Kernel Density Estimation of Features') 
plt.legend() 
plt.show()
```

```python
# pairplot to see relationshipbetween variables , grouped by species
sns.pairplot(data, hue = 'species')
plt.suptitle('Pairplot of Iris data', y = 1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
corr= data.drop('species', axis=1).corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation heatmap of iris data')
plt.show()
```

```python
# Group by species and calculate mean, min etc
grouped = data.groupby('species').agg(['mean', 'std', 'min', 'max'])
grouped
```
